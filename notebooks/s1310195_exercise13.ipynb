## Task1.d: Visualization

i) Write a Python code that read the frequentPatterns.txt file, identifies the longest pattern, and shows it on a plotly express open street map.

```python
import plotly.express as px
import pandas as pd
import re

# Read the file
with open('frequentPatterns.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()

latitudes = []
longitudes = []
frequencies = []

# Process each line
for line in lines:
    line = line.strip()

    # Extract frequency from the end of the line (format: :number)
    freq_match = re.search(r':(\d+)$', line)
    if not freq_match:
        continue  # Skip invalid lines

    frequency = int(freq_match.group(1))

    # Find all POINT(lon lat) occurrences
    points = re.findall(r'POINT\(([-\d\.]+)\s+([-\d\.]+)\)', line)
    for lon, lat in points:
        try:
            longitudes.append(float(lon))
            latitudes.append(float(lat))
            frequencies.append(frequency)
        except ValueError:
            print(f"Skipping invalid coordinates: {lon}, {lat}")

# Create DataFrame and plot
if latitudes and longitudes:
    df = pd.DataFrame({
        'Latitude': latitudes,
        'Longitude': longitudes,
        'Frequency': frequencies
    })

    fig = px.scatter_mapbox(df,
                            lat="Latitude",
                            lon="Longitude",
                            size="Frequency",
                            color="Frequency",
                            color_continuous_scale=px.colors.sequential.Plasma,
                            zoom=4,
                            center={"lat": 36.2048, "lon": 138.2529},
                            height=600)

    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0},
                      coloraxis_colorbar=dict(title="Frequency"))
    fig.show()
else:
    print("No valid location data was found.")
```
